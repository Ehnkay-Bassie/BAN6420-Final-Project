```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load the CSV data into a DataFrame
file_path = 'survey_data.csv'
df = pd.read_csv(file_path)
# Display the first few rows of the DataFrame
df.head()
```

```python
# Group by age and calculate the average income to see the highest income by age
age_income = df.groupby('age')['income'].mean().sort_values(ascending=False)

# Plot the bar chart
plt.figure(figsize=(10, 6))
age_income.plot(kind='bar')
plt.title('Highest Income by Age')
plt.xlabel('Age')
plt.ylabel('Average Income')
plt.tight_layout()
plt.savefig('highest_income_by_age.png')  # Export the chart for use in a PowerPoint presentation
plt.show()
```

## 'Highest Income by Age' Chart Interpretation:
The bar chart illustrates the average income (in millions) for different age groups, displaying from the highest to the lowest. 
Individuals aged 49 and 50 are observed to have the highest average incomes.
Those aged 47 also have a relatively high income, coming in as the second highest.
The trend implies increased earning potential with age, peaking around the late 40s and early 50s, making these age groups particularly attractive targets for a healthcare product launch. 

```python
# Define the spending categories
spending_categories = ['utilities', 'entertainment', 'school_fees', 'shopping', 'healthcare']

# Pivot the DataFrame to have spending categories as rows and genders as columns
df_pivot = df.melt(id_vars=['gender'], value_vars=spending_categories,
                   var_name='spending_category', value_name='amount')

# Aggregate total spending for each spending category by gender
df_grouped = df_pivot.groupby(['spending_category', 'gender']).sum().unstack().fillna(0)

# Plot grouped bar chart
df_grouped.plot(kind='bar', figsize=(12, 8))

# Customize the plot
plt.title('Gender Distribution Across Spending Categories')
plt.xlabel('Spending Category')
plt.ylabel('Total Spending')
plt.xticks(rotation=45)  # Rotate x-axis labels for better readability
plt.legend(title='Gender')
plt.tight_layout()

# Export the chart for use in a PowerPoint presentation
plt.savefig('gender_across_spending_category.png')

# Show the plot
plt.show()
```

## 'Gender Distribution Across Spending Categories' Chart Interpretation
The bar chart illustrates the spending distribution by gender across five categories: entertainment, healthcare, school fees, shopping, and utilities. 
Males exhibit higher spending on entertainment, healthcare, and school fees, with a notable peak in school fees spending. 
In contrast, females spend more on shopping and utilities, with shopping showing the highest spending among females. 
The significant difference in healthcare spending, with males spending more, highlights a key target segment for the new healthcare product launch. 
This analysis helps in tailoring marketing strategies to better reach and engage each gender based on their spending habits.
